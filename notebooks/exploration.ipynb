```python
# Quick sanity check — forward pass through TacticalNet
from torch_geometric.data import Data

model = TacticalNet(player_feature_dim=64, hidden_dim=128, style_latent_dim=4)

def dummy_graph(n=11, feat=64):
    x = torch.randn(n, feat)
    src = [i for i in range(n) for j in range(n) if i != j]
    dst = [j for i in range(n) for j in range(n) if i != j]
    return Data(x=x, edge_index=torch.tensor([src, dst]), batch=torch.zeros(n, dtype=torch.long))

da, db = dummy_graph(), dummy_graph()
sa = torch.randn(1, 4)
sb = torch.randn(1, 4)
logits = model(da, db, sa, sb)
print('Output logits:', logits)
print('Probabilities:', torch.softmax(logits, dim=-1))
```
